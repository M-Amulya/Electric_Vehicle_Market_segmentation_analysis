### **EV-Geographical analysis**

Analysing the geographical electric vehicle data to identify the target region.

```python
#load the data
import pandas as pd
df3=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/EV_market_segmentation/Geographical_dataset.csv')
```

```python
df3.head()
```

```python
#check shape
df3.shape
```

```python
#check basic descriptive statistics
df3.describe()
```

```python
# check basic info of dataset
df3.info()
```

```python
#Ploting the region wise two wheeler count
import matplotlib.pyplot as plt
import seaborn as sns
ax= plt.figure(figsize=(10,2))
sns.barplot(x='Region', y='2W', data=df3)
plt.ylabel('2W')
plt.xlabel('Region')
plt.title('Usage of two wheelers in each Region')
plt.xticks(rotation=90)
plt.show()
```

From above plot it is clear that most of the two wheeler electric vehicles are used in the states of Maharastra, Karnataka,Tamilnadu

```python
#Ploting the region wise three wheeler count
import matplotlib.pyplot as plt
import seaborn as sns
ax= plt.figure(figsize=(10,2))
sns.barplot(x='Region', y='3W', data=df3)
plt.ylabel('3W')
plt.xlabel('Region')
plt.title('Usage of three wheelers in each Region')
plt.xticks(rotation=90)
plt.show()
```

From above plot it is clear that most of the three wheeler electric vehicles are used in Uttar Pradesh

```python
#Ploting the region wise 4 wheeler count
import matplotlib.pyplot as plt
import seaborn as sns
ax= plt.figure(figsize=(10,2))
sns.barplot(x='Region', y='4W', data=df3)
plt.ylabel('4W')
plt.xlabel('Region')
plt.title('Usage of four wheelers in each Region')
plt.xticks(rotation=90)
plt.show()
```

Most of the 4 wheeler electric vehicles are used in the states of Maharastra and Delhi.

```python
#Ploting the region wise electric buses count
import matplotlib.pyplot as plt
import seaborn as sns
ax= plt.figure(figsize=(10,2))
sns.barplot(x='Region', y='Bus', data=df3)
plt.ylabel('Bus')
plt.xlabel('Region')
plt.title('Usage of Buses in each Region')
plt.xticks(rotation=90)
plt.show()
```

```python
#Plotting the usage of chargers in each region
import matplotlib.pyplot as plt
import seaborn as sns
ax= plt.figure(figsize=(10,2))
sns.barplot(x='Region', y='Chargers', data=df3)
plt.ylabel('Chargers')
plt.xlabel('Region')
plt.title('Usage of Chargers in each Region')
plt.xticks(rotation=90)
plt.show()
```

**Conclusion:**
Most of the 4 wheeler electric vehicles are used in the states of Maharastra and Delhi.
We can conclude that this states are good to sell our EV cars,
So our targeted states are Maharastra and Delhi.
